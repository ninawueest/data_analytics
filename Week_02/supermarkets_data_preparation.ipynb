# Importing and preparing supermarkets data

## Libraries and settings

```python
# Libraries
import os
import fnmatch
import pandas as pd

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
print('Current working directory:', os.getcwd())

# Show .json files in the current working directory
flist = fnmatch.filter(os.listdir('.'), '*.json')
for i in flist:
    print(i)
```

## Importing data

```python
# Read the data to a pandas data frame
df1 = pd.read_json('supermarkets.json', encoding='utf-8')
df1.head(5)
```

## Count number of rows and columns in the data frame

```python
# Dimension (rows, columns)
print('Dimension:', df1.shape)

# Number of rows
print('Number of rows:', df1.shape[0])

# Number of columns
print('Number of columns:', df1.shape[1])
```

## Column 'tags' is a pandas Series with dictionaries -> change to data frame

```python
# Type of the first item of column tags
print(type(df1.tags))
print(type(df1.tags[0]))

# Content of the first item of column tags
print(df1.tags[0].keys())

# Change to data frame
df2 = pd.DataFrame.from_records(df1.tags)
df2[['brand', 'shop', 'addr:city', 'addr:street', 'addr:housenumber', 'addr:postcode']]

# Rename selected columns
df2 = df2.rename(columns={'addr:city': 'city',
                          'addr:street':'street',
                          'addr:housenumber': 'housenumber',
                          'addr:postcode': 'postcode'})

# Show first records of data frame
df2.head()
```

## Merge df1 and df2

```python
# Merge df and df2
df = pd.merge(df1[['type', 'id', 'lat', 'lon']], 
              df2[['brand', 'shop', 'city', 'street', 'housenumber', 'postcode']],
              left_index=True, 
              right_index=True)
df.head(5)
```

## Count and identify the number of missing values (if any)

```python
# Count missing values
print(pd.isna(df).sum())

# Identify rows with missing values, e.g.:
df.loc[pd.isna(df['city'])]
```

## Count and identify duplicated values (if any)

```python
# Count duplicated values
print(df.duplicated().sum())

# Identify rows with duplicated values, e.g.:
df[df[['id']].duplicated()]
```

## Get data types of all variables

```python
# Get data types (note that in pandas, a string is referred to as 'object')
df.dtypes
```

### Save data to file

```python
df.to_csv('supermarkets_data_prepared.csv', 
          sep=",", 
          encoding='utf-8',
          index=False)
```

## Additional filters on supermarkets

<h3>Filter only Migros supermarkets in the city of Zürich</h3>

```python
# Filter only Migros supermarkets in the city of Zürich.
df_filtered = df.loc[(df['brand'] == 'Migros') & (df['city'] == 'Zürich')]
df_filtered
```

<h3>Filter and count all Coop supermarkets in the cities of Zürich, Basel & Bern</h3>

```python
# Filter and count all Coop supermarkets in the cities of Zürich, Basel & Bern.
cities = ['Zürich', 'Basel', 'Bern']
df_filtered = df.loc[(df['brand'] == 'Coop') & (df['city'].isin(cities))]

# Count the number of rows that match the criteria
count = df_filtered.shape[0]  
print("Count of all Coop supermarkets in the cities Zürich, Basel, and Bern: ", count)
df_filtered
```

<h3>Filter supermarkets with available brand, city, house number and postcode</h3>

```python
# Filter supermarkets with available brand, city, house number and postcode.
df_filtered = df.loc[(df['brand'].notna()) & (df['city'].notna()) & (df['housenumber'].notna()) & (df['postcode'].notna())]
df_filtered
```
