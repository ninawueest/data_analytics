# Exploratory data analysis (EDA) of supermarket data

## Libraries and settings

```python
# Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn graphics settings
sns.set(color_codes=True)

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Read CSV File
df = pd.read_csv('supermarkets_data_enriched.csv')
```

## Count the number of supermarkets per brand

```python
brand_count = df['brand'].value_counts()
brand_count
```

## Create a barchart with number of supermarkets per brand

```python
plt.figure(figsize=(10, 7))
sns.barplot(x=brand_count.index, y=brand_count.values)
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Number of supermarkets')
plt.title('Number of supermarkets per brand')
plt.show()
```

## Create a scatterplotmatrix of numeric variables

```python
g = sns.PairGrid(df, vars=['lat', 'lon', 'pop', 'pop_dens', 'frg_pct', 'emp'])
g.map(plt.scatter)
plt.show()
```

## Create a plot with the locations uf supermarkets in different colors acc. to brand

```python
# Define the brands to plot
brands_plot = ['Denner', 'Volg', 'Landi', 'Coop']

# Filter the dataframe for the selected brands
filtered_df = df[df['brand'].isin(brands_plot)]

# Create a scatterplot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=filtered_df, x='lon', y='lat', hue='brand', palette='Set1', s=100)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Supermarket Locations by Brand')
plt.legend(title='Brand', loc='upper right')
plt.show()
```
